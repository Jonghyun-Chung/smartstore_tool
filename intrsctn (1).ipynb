```python
#dur은 str기간, age는 int나이(10대 ->1, 20대->2...), yr_start는 int, device는 "pc"나 "mov"
def byAgeYr(dur, age, yr_start, device, lst, new_lst):
    termdiv = driver.find_element_by_class_name("select.w4")
    termdiv.find_element_by_class_name("select_btn").click()
    termdiv.find_element_by_link_text(dur).click()
    monthdiv = driver.find_elements_by_class_name("set_period")[1]
    monthdiv.find_elements_by_class_name("period")[3].click()
    first_termdiv = driver.find_element_by_class_name("select.w2")
    first_termdiv.find_element_by_class_name("select_btn").click()
    first_termdiv.find_element_by_link_text(str(yr_start)).click()
    first_monthdiv = driver.find_element_by_class_name("select.w3")
    first_monthdiv.find_element_by_class_name("select_btn").click()
    first_monthdiv.find_element_by_link_text("12").click()
    final_termdiv = driver.find_elements_by_class_name("select.w2")[1]
    final_termdiv.find_element_by_class_name("select_btn").click()
    final_termdiv.find_element_by_link_text(str(int(yr_start)+1)).click()
    final_monthdiv = driver.find_elements_by_class_name("select.w3")[2]
    final_monthdiv.find_element_by_class_name("select_btn").click()
    final_monthdiv.find_element_by_link_text("01").click()
    if device == "pc":
        ck = 2
    elif device == "mov":
        ck = 3
    driver.find_elements_by_class_name("chk")[ck].click()
    
    driver.find_element_by_id("19_gender_1").click()
    driver.find_element_by_id("20_age_" + str(age)).click()
    driver.find_element_by_class_name('btn_submit').click()
    
    for i in range(25):
        driver.find_element_by_class_name('btn_page_next').click()
        clo_html = driver.page_source
        clo_html = bs(clo_html, 'html.parser')
        clo_item = clo_html.find_all("div", {"class":"rank_top1000"})
        lst += clo_item[0].text.split()
    for i in lst:
        x = re.sub('[^a-zA-Zㄱ-힗]', '', i)
        if x!= "":
            new_lst.append(x)
    
    final_termdiv.find_element_by_class_name("select_btn").click()
    final_termdiv.find_element_by_link_text("2020").click()
            
    driver.find_elements_by_class_name("chk")[0].click()
    driver.find_elements_by_class_name("chk")[0].click()

    driver.find_element_by_id("19_gender_0").click()
    driver.find_element_by_id("19_gender_0").click()

    driver.find_element_by_id("20_age_0").click()
    driver.find_element_by_id("20_age_0").click()
```

```python
def intr2(lst18, lst19):
    return list(set(lst18) & set(lst19))
```

```python
def intr3(lst17, lst18, lst19):
    return list(set(lst17) & set(lst18) & set(lst19))
```

```python

inter_30_8 = intr3(b30_8_17, b30_8_18, b30_8_19)
inter_40_8 = intr3(b40_8_17, b40_8_18, b40_8_19)

inter_30_7 = intr2(b30_7_18, b30_7_19)
inter_40_7 = intr2(b40_7_18, b40_7_19)

print(inter_30_8)
print("//////////////////")
print(inter_40_8)
print("//////////////////")
print(inter_30_7)
print("//////////////////")
print(inter_30_8)
```
